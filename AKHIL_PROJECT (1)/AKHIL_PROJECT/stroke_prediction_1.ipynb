## Bivariate Analysis

### Analysing relationship between target variable(Diagnosis) and reamining features

```python

# Define the target column
target_column = "Diagnosis"

# Define the numerical and categorical columns
numerical_columns = ['Age', 'Average Glucose Level', 'Body Mass Index (BMI)', 'Stress Levels', 'Cholestrol levles HDL', 'Cholestrol levles LDL']
categorical_columns = ['Gender', 'Hypertension', 'Heart Disease', 'Marital Status', 'Smoking Status', 'Alcohol Intake', 'Physical Activity',
                      'Stroke History', 'Family History of Stroke', 'Dietary Habits']

# Iterate through the columns and create the appropriate plots
for column in df.columns:
    if column != target_column:
        if column in numerical_columns:
            # Scatter plot for numerical vs. numerical
            sns.scatterplot(data=df, x=column, y=target_column)
            plt.title(f'{column} vs. {target_column}')
            plt.show()
        elif column in categorical_columns:
            # Bar plot for categorical vs. categorical
            sns.countplot(data=df, x=column, hue=target_column)
            plt.title(f'{column} vs. {target_column}')
            plt.show()
        else:
            # Boxplot for numerical vs. categorical
            sns.boxplot(data=df, x=target_column, y=column)
            plt.title(f'{column} vs. {target_column}')
            plt.show()
```

```python
def models(hp):
    model=Sequential()
    model.add(Input(shape=(26,)))
    for i in range(hp.Int("hidden_layers",min_value=10,max_value=300,step=10)):
    
        model.add(Dense(units=hp.Int('neurons{}'.format(i),min_value=4,max_value=200,step=1),
                   activation=hp.Choice('activation',['relu','tanh','sigmoid'])))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(
    optimizer=hp.Choice("optimizer",['sgd','adam','adadelta','rmsprop']),loss='binary_crossentropy',metrics=['accuracy'])
    
    return model
```
